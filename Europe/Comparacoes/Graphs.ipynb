```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r"C:\Users\Alysson\Documents\GitHub\Monetary-Shocks\Europe\Comparacoes\taxas de juros.xlsx")
df.head()
```

```python
# Plotando o gráfico de linhas
plt.figure(figsize=(10, 6))

plt.plot(df['Date'], df['US'], label='US', color='navy')
plt.plot(df['Date'], df['Europe'], label='Europe', color='cornflowerblue')
plt.plot(df['Date'], df['Brazil'], label='Brazil', color='lightsteelblue')


plt.xlabel('Year')
plt.ylabel('Interest Rates (%)')
plt.legend()

# Exportar o gráfico como um arquivo PNG
plt.savefig('historico.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
# Defina as variáveis independentes e dependentes
X = df['US']
y = df['Brazil']

# Adicione uma constante ao conjunto de dados para o termo intercepto
X = sm.add_constant(X)

# Crie um modelo OLS
model = sm.OLS(y, X)

# Ajuste o modelo
results = model.fit()

# Imprima os resultados da regressão
print(results.summary())
```
