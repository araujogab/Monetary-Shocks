```python
from scipy.optimize import fmin
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

path = r"C:\Users\Alysson\Documents\GitHub\Monetary-Shocks\Brasil\ETTJ\Dados_teste.xlsx"
df = pd.read_excel(path)
df.head()
```

```python
def myval(c):    
    df['NSS'] =(c[0])+(c[1]*((1-np.exp(-df['Maturity']*0.02))/(df['Maturity']*0.02)))+(c[2]*((((1-np.exp(-df['Maturity']*0.02))/(df['Maturity']*0.02)))-(np.exp(-df['Maturity']*0.02))))+(c[3]*((((1-np.exp(-df['Maturity']*0.20))/(df['Maturity']*0.20)))-(np.exp(-df['Maturity']*0.20))))
    discount = []
    for index, row in df.iterrows():
        discount.append((row['Coupon']/(1+df['NSS'])**df['Maturity'])[0:index+1].sum())
    df['Discount'] = discount
    df['Calculated_price'] = 100 / (1 + df['NSS']) ** df['Maturity'] + df['Discount']  
    df['Residual'] =  (df['Price'] - df['Calculated_price'])**2
    val = np.sum(df['Residual'])
    print("[b1, b1, b2, b3, lda1, lda2]=",c,", SUM:", val)
    return(val)
    
c = fmin(myval, [0.05, 0.0, 0.0, 0.0])
```
