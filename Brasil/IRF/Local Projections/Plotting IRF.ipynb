```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel("IRF DF1.xlsx")
df.head()
```

```python


# Configurando subplots
fig, axs = plt.subplots(2, 1, figsize=(4, 5))

list_var = ["log_ipca", "log_pimsa"]

for i, variable in enumerate(list_var):
    # Plotando as IRFs no subplot correspondente
    ax = axs[i]

    # Recuperando as listas correspondentes  varivel do DataFrame
    LP_b_lhs_var = df[f'LP_b_lhs_{variable}']
    LP_se_lhs_var = df[f'LP_se_lhs_{variable}']

    # Plotando a linha central
    ax.plot(df['h'], LP_b_lhs_var, label='IRF', color='DarkSeaGreen', linewidth=2)

    # Preenchendo a rea entre o intervalo de confiana
    ax.fill_between(df['h'], np.array(LP_b_lhs_var) - 1.96 * np.array(LP_se_lhs_var),
                    np.array(LP_b_lhs_var) + 1.96 * np.array(LP_se_lhs_var), color='gray', alpha=0.3, label='95% CI')

    # Adicionando uma linha em y = 0
    ax.axhline(0, color='black', linestyle='--', linewidth=1)

    # Adicionando rtulos e ttulos
    ax.set_title(f'{variable}')
    ax.set_xlabel('Month')
    # ax.set_ylabel(f'Cumulated change in log {variable}')

    # Adicionando uma legenda
    ax.legend()

# Ajustando o layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
axs[0].set_title('IPCA')
axs[1].set_title('Industrial Production')
#axs[2].set_title('Financial Conditions')
#axs[3].set_title('BRL/EM exchange rates')


# Exportar o grfico como um arquivo PNG
plt.savefig('DF1_IRF_vert.png', dpi=300, bbox_inches='tight')

# Exibindo os subplots
plt.show()
```
